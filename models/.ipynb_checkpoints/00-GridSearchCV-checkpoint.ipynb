```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
df = pd.read_csv('advertising.csv')
df
```

```python
df = pd.get_dummies(df, columns=['City', 'Country'], drop_first=True)
df
```

```python
x = df.drop(['Ad Topic Line', 'Timestamp', 'Clicked on Ad'], axis=1)
y = df['Clicked on Ad']
```

```python
model = SVC()
```

```python
from sklearn.model_selection import GridSearchCV

params = {'C' : [1,5,7], 'kernel' : ['linear', 'poly', 'rbf']}

grid_search = GridSearchCV(model, param_grid=params, scoring = 'accuracy', cv = 5)

grid_search.fit(x, y)
```

```python
grid_search.best_params_
```

```python
grid_search.cv_results_
```

### More than scoring function

```python
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score

params = {'C' : [1,5,7], 'kernel' : ['linear', 'poly', 'rbf']}

scoring = {'accuracy' : make_scorer(accuracy_score),
          'precision' : make_scorer(precision_score, average = 'macro'),
          'recall' : make_scorer(recall_score, average = 'macro')}

grid_search = GridSearchCV(model, param_grid=params, scoring = scoring, cv = 5, refit = False)

grid_search.fit(x, y)
```

```python
grid_search.cv_results_
```
