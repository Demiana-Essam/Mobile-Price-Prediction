```python
import numpy as np
import pandas as pd
import matplotlib.pyplot
import seaborn as sns
sns.set(rc={'figure.figsize': [15, 5]}, font_scale=1.3)
```

```python
the_data = pd.read_csv('bikes.csv')
the_data
```

```python
the_data.describe()

```

```python
the_data.info()
```

# 1. Make Profit Feature.

```python
def profit_feat(the_data):
    prof = the_data['casual']*20 + the_data['registered']*5 
    prof -= prof*0.14/100
    return  prof

```

```python
the_data['profit']= the_data.apply(profit_feat , axis=1)
the_data.head()
```

# 2. Distribution rental_bikes_count & profit.

```python
histo = the_data[['rented_bikes_count','profit']].hist(bins=10)
```

# 3. Profit for each day.

```python
the_data['date'] = pd.to_datetime(the_data['datetime'].str.split(" ",expand=True)[0])

the_data[['hours','min','sec']] = the_data['datetime'].str.split(" ",expand=True)[1].str.split(":",expand=True)
the_data['hours'] = pd.to_numeric(the_data['hours'],downcast="integer")
the_data.drop('datetime',axis=1,inplace=True)

the_data
```

```python

the_data.groupby('date').sum()['profit'][:30].plot(kind='bar')
```

# 4. Which season that has more registration & profit.

```python
the_data.groupby('season',as_index=False).sum()['season'][0]
```

```python
sns.boxplot(data = the_data, x = 'season', y = 'profit')
```

# 5. Which weather condition that has more registration & profit.

```python
the_data.groupby('weather',as_index=False).sum()['weather'][0]
```

```python
sns.boxplot(data = the_data, x = 'weather', y = 'profit')
```

# 6. Correlation between profit & bikes count with all features.

```python
the_data['sum_of_profit_and_bikes_count'] = (the_data['rented_bikes_count'] + the_data['profit'])
the_data.corr()['sum_of_profit_and_bikes_count']
```

```python
sns.heatmap(the_data.corr(), annot=True, cmap = 'Blues')
```

# 7. Rentals during rush hours (7-9 am & 3-5 pm).

```python
d=the_data[the_data['hours'].between(7, 9) | the_data['hours'].between(15, 17)][['hours','rented_bikes_count']]
```

```python
sns.boxplot(y='rented_bikes_count', x='hours', data=d)
```

# 7. Rentals during workday & holiday.

```python
the_data[(the_data['workingday']==1) | (the_data['holiday']==1)][['holiday','workingday','rented_bikes_count']]
```

```python
sns.boxplot(y='rented_bikes_count', x='workingday',hue = 'holiday', data=the_data)
```

# 8. Registered or casual is more.

```python
if the_data['registered'].sum() > the_data['casual'].sum():
    print ('registered is more ')
else:
    print('casual is more')
```

# 9. What is the average of bikes would be registered during the week.

```python
the_data.groupby(pd.Grouper(key='date',  freq='7D')).mean()['registered'].head(30).plot(kind = 'bar')
```

# 10. Schools (9-12) & (2-6) profit.

```python
the_data[the_data['hours'].between(9, 12) | the_data['hours'].between(2, 6)][['hours','profit']].groupby('hours').mean().plot(kind = 'bar')
```

# 11. Profit in 2011 & Saturdays & season fall

```python
d = the_data[(the_data['date'].dt.day_name() == 'Saturday')  & (the_data['date'].dt.year == 2011) & (the_data['season'] == 'Fall')]
d[['date','season','profit']]
```

```python
d.groupby('date').sum()['profit'].plot(kind = 'bar')
```
