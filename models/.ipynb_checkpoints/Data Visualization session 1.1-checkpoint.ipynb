# Distribution Plots

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* 1) Univariate Plots
    * 1. distplot
    * 2. kdeplot
* 2) Bivariate Plots
    * 1. jointplot
    * 2. pairplot

## Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set(rc={'figure.figsize': [8, 8]}, font_scale=1.2)
```

**seaborn styles**
- darkgrid, whitegrid, dark, white and ticks

```python
# sns.set_style('darkgrid')
```

## Data
Seaborn comes with built-in data sets!

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips['day'].value_counts()
```

## 1) Univariate Plots
plots that deal with one variable
* 1. Histogram
* 2. kdeplot
* 3. countplot

### 1. Histogram

The distplot shows the distribution of a univariate set of observations.

```python
sns.histplot(data = tips, x = 'total_bill', bins = 20, kde = True)
plt.title('Total Bill Distribution')
```

Change the color a little bit, valid options for colors: r, g, b, c, m, y, k

```python
sns.histplot(data = tips, x = 'tip', bins = 20, kde = True)
```

### 2. kdeplot

kdeplots are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value. For example:

```python
sns.kdeplot(data = tips, x = 'tip')
```

### 3. CountPlot

```python
## used for categorical feature
sns.countplot(data = tips, x = 'day')
```

```python
color = sns.color_palette()[-1]
```

```python
idx = tips['size'].value_counts().index
```

```python
sns.countplot(data = tips, x = 'size', color = color, order = idx)
```

## 2) Bivariate Plots
plots that deal with two variables
* 1. scatterplot
* 2. jointplot
* 3. pairplot
* 4. Boxplot
* 5. Violinplot
* 6. Barplot

### 1. scatterplot



```python
tips.head()
```

```python
## scater plot is used to get the relation between two quantitative columns
sns.scatterplot(data = tips, x = 'tip', y = 'total_bill')
```

### 2. Jointplot

```python
sns.jointplot(y='total_bill', x='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', height=8, color='m')
```

```python
# sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', height=8, color='m')
```

### 3. pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
sns.pairplot(tips)
```

Change the size a little bit

```python
sns.pairplot(tips, height=3.5, aspect=1.3)
```

Choose custom columns

```python
sns.pairplot(tips, vars=['total_bill', 'tip'], height=5, aspect=1.3)
```

Use hue to show graph based on the hue category values

```python
# sns.pairplot(tips, hue='sex', height=3, aspect=1.3)
```

```python
# sns.pairplot(tips, hue='day', height=3, aspect=1.3)
```

```python
# sns.pairplot(tips, hue='time', height=3, aspect=1.3)
```

### 4. Boxplot

```python
sns.boxplot(data = tips, x = 'sex', y = 'total_bill')
```

```python
sns.boxplot(data = tips, x = 'day', y = 'total_bill')
```

```python
tips.head()
```

```python
sns.boxplot(data = tips, x = 'size', y = 'total_bill', palette='viridis')
```

### 5. Violinplot

```python
sns.violinplot(data = tips, x = 'sex', y = 'total_bill', inner = None)
```

```python
sns.violinplot(data = tips, x = 'sex', y = 'tip', inner = None)
```

### 6. Barplot

```python
sns.barplot(data = tips, x = 'sex', y = 'total_bill', ci = None)
```

```python
sns.barplot(data = tips, x = 'size', y = 'total_bill', ci = None, color = color)
```

## Three columns

```python
sns.scatterplot(data = tips, x = 'tip', y = 'total_bill', hue  ='sex')
```

```python
sns.scatterplot(data = tips, x = 'tip', y = 'total_bill', hue  ='smoker')
```

```python
tips.head()
```

```python
import pandas as pd
x = tips[tips['sex'] == 'Male']['smoker']
x =  pd.DataFrame(data = x.values, columns = ['smoker'])
sns.countplot(data = x, x = 'smoker')
```

```python
xx = tips[tips['sex'] == 'Male']['smoker'].value_counts()
xx
sns.barplot(x = xx.index, y = xx.values)
```

```python
sns.countplot(data = tips, x = 'sex', hue = 'smoker')
```

```python
sns.violinplot(data = tips, x = 'sex', y = 'tip', hue = 'smoker', inner = None)
```
